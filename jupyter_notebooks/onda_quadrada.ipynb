# Visualizando uma Onda Quadrada com Matplotlib e NumPy e SciPy

- Ondas Quadradas são formas de onda periódicas
- Entretanto, Ondas Quadradas são não-sinusoidais. A transição entre os valores de pico é instantâneo em uma onda quadrada
- O período da Onda Quadrada é também chamado de largura do pulso
- Para desenhar uma Onda Quadrada utilizando matplotlib, numpy e scipy é necessário alguns detalhes:
    - Frequência da Onda Quadrada, digamos 10Hz, que é 10 ciclos por segundo
    - A frequência de amostragem, ou seja, quandos pontos de dados no qual a onda está sendo construída, quanto mais elevado os pontos de dados, mais 'suave' o quadrado é.
    - Ondas Quadradas possuem um ciclo de trabalho de 50%, ou seja, a porcentagem da forma de onda que ocorre acima do eixo zero é de 50%.
    - Por padrão, a função **signal.square()** da biblioteca scipy recebe um valor de trabalho de 0.5

```python
from scipy import signal
import matplotlib.pyplot as plot
import numpy as np
%matplotlib inline
plot.rcParams["figure.figsize"] = [16,9]

# Taxa de amostragem 1000 hz / segundo
t = np.linspace(0, 1, 1000, endpoint=True)

# Plotamos o sinal da onda quadrada
plot.plot(t, signal.square(2 * np.pi * 5 * t))

# Damos um título a Plotagem
plot.title('Onda Quadrada - 5 Hz amostrado em 1000 Hz / segundo')

# Rotulamos o eixo-X
plot.xlabel('Tempo')

# Rotulamos o eixo-Y
plot.ylabel('Amplitude')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')

# Seta o valor max e min para o eixo Y
plot.ylim(-2, 2)

# Desenha o gráfico da onda quadrada
plot.show()
```
