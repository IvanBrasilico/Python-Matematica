# Função Tangente Hiperbólica

Na Matemática, as funções hiperbólicas são análogas às funções trigonométricas comuns.

As funções hiperbólicas básicas são:

- Seno hiperbólico "sinh"
- Cosseno hiperbólico "cosh"

No qual podemos derivar:

- **Tangente hiperbólica "tanh"**
- Cossecante hiperbólica "cosech"
- Secante hiperbólica "sech"
- Cotangente hiperbólica "coth"

A **tanh** ou **função de ativação tangente hiperbólica** tem a seguinte fórmula matemática:

\begin{equation}
\tanh x=\frac{\cosh x}{\sinh x} = \frac{e^x - e^{-x}}{e^x + e^{-x}}
\end{equation}

Ela basicamente recebe um número de valor real e coloca ele entre $-1$ e $+1$.

Vejamos um exemplo da Função Tanh

```python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline
plt.rcParams["figure.figsize"] = [16,9]

in_array = np.linspace(-np.pi, np.pi, 12) 
out_array = np.tanh(in_array) 
  
print("array de input : ", in_array) 
print("\narray de output: ", out_array) 
  
plt.plot(in_array, out_array, color = 'b') 
plt.title("Função Hiperbólica") 
plt.xlabel("Eixo X") 
plt.ylabel("Eixo Y") 
plt.grid()
plt.show() 
```
