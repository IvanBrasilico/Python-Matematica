# Função de Unidade Linear Retificada

A função de Unidade Linear Retificada, muito conhecida por ReLU (Rectified Linear Unit) é a função de ativação mais utilizada em modelos de Deep Learning. A função retorna 0 se ela receber qualquer input negativo, porém para cada valor positivo $x$ ela retorna o valor de volta. Ela pode ser escrita da seguinte maneira:

\begin{equation}
f(x) = max(0,x)
\end{equation}

O exemplo abaixo gera uma série de inteiros entre **-10** até **10** e calcula a ativação da unidade linear retificada para cada input

```python
from matplotlib import pyplot
import matplotlib as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = [16,9]
 
# Função de Unidade Linear Retificada
def rectified(x):
    return max(0.0, x)
 
# Define a série de inputs
series_in = [x for x in range(-10, 11)]
# Cálcula os outputs para nossos inputs
series_out = [rectified(x) for x in series_in]
print(series_in)
print(series_out)
pyplot.plot(series_in, series_out)
pyplot.show()
```
