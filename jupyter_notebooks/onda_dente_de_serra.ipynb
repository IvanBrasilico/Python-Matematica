# Visualizando uma Onda Dente de Serra com Matplotlib, NumPy e SciPy

- Uma Onda Dente de Serra é uma forma de onda periódica e ela é não-sinusoidal
- Uma Onda Dente de Serra cresce para cima e cai bruscamente. Uma Onda Dente de Serra também pode ir para baixo e ascender bruscamente, neste caso ela é chamada de "onda dente de serra reversa" ou "onda dente de serra inversa"
- Ondas Dente de Serra tem suas aplicações em sintetizadores musicais, em telas de Tubo de Raios Catódicos (CRT) e Osciloscópios.
- A função **signal.sawtooth()** retorna uma forma de onda dente serra periódica ou uma forma de onda triangular

```python
# Importamos as bibliotecas necessárias
import numpy as np
from scipy import signal
import matplotlib.pyplot as plot
%matplotlib inline
plot.rcParams["figure.figsize"] = [16,9]

# Criamos 1000 pontos separados linearmente com valores entre 0 e 1
pontos_temporais = np.linspace(0, 1, 500)

# Frequência da Onda Dente de Serra 5Hz. Frequência da amostragem 500 Hz
plot.plot(pontos_temporais, signal.sawtooth(2 * np.pi * 5 * pontos_temporais))

# Damos um título ao Gráfico
plot.title('Onda Dente de Serra - 5 Hz amostrado em 1000 Hz / segundo')

# Rotulamos o eixo-X
plot.xlabel('Tempo')

# Rotulamos o eixo-Y
plot.ylabel('Amplitude')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')

# Seta o valor max e min para o eixo Y
plot.ylim(-2, 2)

# Desenha o gráfico da onda dente de serra
plot.show()
```
