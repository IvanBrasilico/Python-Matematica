# Visualizando uma Onda Cossenoidal com Matplotlib e NumPy

- Ondas Cossenoidais são ondas periódicas geradas através de oscilações
- Ondas Cossenoidais são similares às ondas senoidais, entretanto a onda cossenoidal lidera a onda senoidal por um ângulo fase de 90 graus
- A curva do cosseno não passa pela origem (0,0)
- A onda cossenoidal tem a mesma frequência de sua correspondente onda senoidal
- As ondas marítmas são um exemplo de ondas cossenoidais
- Uma curva de cosseno pode ser plotada ao chamarmos a função cosine() em um array NumPy e utilizarmos a função plot() da biblioteca matplotlib

Veja um exemplo

```python
import numpy as np
import matplotlib.pyplot as plot
%matplotlib inline
plot.rcParams["figure.figsize"] = [16,9]
 
# Obtém os valores X da onda cossenoidal
tempo = np.arange(0, 10, 0.2);

# A amplitude da onda cossenoidal é o seno de uma variável, nesse caso será tempo
amplitude = np.cos(tempo)

# Plotamos a onda cossenoidal utilizando tempo e amplitude
plot.plot(tempo, amplitude)

# Damos um título para a plotagem
plot.title('Onda Cossenoidal')

# Rotulamos o eixo-X
plot.xlabel('Tempo')

# Rotulamos o eixo-Y
plot.ylabel('Amplitude = cosine(tempo)')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')

# Mostra o gráfico da onda cossenoidal
plot.show()
```
