# Visualizando Gráficos com SymPy

Importando os módulos necessários

```python
from IPython.display import display, Math
from sympy import *
init_printing()
```

## Plotagem

Função de única variável

```python
from sympy import symbols
from sympy.plotting import plot
%matplotlib inline

x = symbols('x')

expr = x*x

plot(expr);
```

## Plotagem com alcance

Para especificarmos o alcance da plotagem, passamos uma tupla na forma (**variável**, **limite_menor**, **limite_maior**)

```python
expr = x**2
expr_range = (x,-3,7)

plot(expr, expr_range);
```

## Título e Rótulo

Passamos os argumentos com as palavras-chave **Title**, **xlabel**, **ylabel**. LateX é suportado por esses argumentos.

```python
expr = x**2
expr_range = (x,-2,3)

title = '$y = {}$'.format(latex(expr))

plot(expr, expr_range, title = title, xlabel = 'x', ylabel = 'y');
```

## Cor da Linha

Passamos o argumento **line_color** para definirmos a cor da linha

```python
x = symbols('x')

expr = x**2

plot(expr, expr_range, line_color = 'g');
```

## Múltiplas Plotagens com o mesmo alcance

Para isso utilizamos a sintáxe **plot(expr_1, expr_2, expr_3, alcance)**

```python
expr_1 = x
expr_2 = x**2
expr_3 = x**3

plot(expr_1, expr_2, expr_3, (x, -1, 1));
```

## Múltiplas Plotagens com diferentes alcances

Para isso, utilizamos a sintáxe:

plot(
    (expr_1,range_1),
    (expr_2,range_2),
    ...
)

```python
expr_1 = x**2
range_1 = (x,-2,2)

expr_2 = x
range_2 = (x,-1,3)

plot(
    (expr_1,range_1),
    (expr_2,range_2)
);
```

## Múltiplas Plotagens com cores diferentes

1. Passamos **show = False** para suprimirmos a exibição do gráfico, salvamos o objeto retornado para uma variável
2. Obtemos as série de dados individuais através da indexação do objeto plotting e setamos a **line_color** separadamente
3. Chamamos o método **show()** do objeto plotting para mostrarmos a imagem

```python
expr_1 = x**2
range_1 = (x,-2,2)

expr_2 = x
range_2 = (x,-1,1)

p = plot(
    (expr_1,range_1),
    (expr_2,range_2),
    show = False
);

p[0].line_color = 'g'
p[1].line_color = 'r'

p.show()
```

Para adicionarmos uma legenda ao Gráfico, é necessário dois passos adicionais

1. Passamos **Legend = True** no momento de construção do objeto plotting
2. Setamos um rótulo para cada série de dados, separadamente

```python
expr_1 = x**2
range_1 = (x,-2,2)

expr_2 = x
range_2 = (x,-1,1)

p = plot(
    (expr_1,range_1),
    (expr_2,range_2),
    show = False,
    legend = True
);

p[0].line_color = 'g'
p[1].line_color = 'r'

p[0].label = 'Linha 1'
p[1].label = 'Linha 2'

p.show()
```

## Gráfico 3D

Plotagem única

Utilizamos a função **plot3d(expr, alcance_x, alcance_y)** para desenharmos um gráfico 3D

```python
from sympy import symbols
from sympy.plotting import plot3d

x, y = symbols('x y')

expr = x*y
alcance_x = (x, -5, 5)
alcance_y = (y, -5, 5)

plot3d(expr, alcance_x, alcance_y);
```

## Múltiplas plotagens

Para desenharmos múltiplas plotagens de um gráfico 3D, utilizamos a seguinte sintáxe

plot3d(
    （expr_1, alcance_x_1 , alcance_y_1）,
    （expr_2, alcance_x_2 , alcance_y_2）
)

```python
plot3d(
    (x**2 + y**2, (x, -5, 5), (y, -5, 5)),
    (x*y, (x, -3, 3), (y, -3, 3))
);
```

## Função paramétrica de variável única

Utilizamos a função **plot_parametric(expr_x, expr_y, alcance_u)** para desenharmos uma função paramétrica de variável única

```python
from sympy import symbols, cos, sin
from sympy.plotting import plot_parametric

u = symbols('u')
expr_x = cos(u)
expr_y = sin(u)

p = plot_parametric(expr_x, expr_y, (u, -5, 5));
```

## Linha paramétrica 3D

Utilizamos a função **plot3d_parametric_line(expr_x, expr_y, expr_z, alcance_u)** para desenharmos uma função paramétrica 3D

```python
from sympy import symbols, cos, sin
from sympy.plotting import plot3d_parametric_line
u = symbols('u')

expr_x = cos(u)
expr_y = sin(u)
expr_z = u

plot3d_parametric_line(expr_x, expr_y, expr_z, (u, -5, 5));
```

## Superfície paramétrica 3D

Utilizamos a função **plot3d_parametric_surface(expr_x, expr_y, expr_z, alcance_u, alcance_v)** para desenharmos uma superfície 3D paramétrica

```python
from sympy import symbols, cos, sin
from sympy.plotting import plot3d_parametric_surface
u, v = symbols('u v')

expr_x = cos(u + v)
expr_y = sin(u-v)
expr_z = u-v
u_range = (u, -5, 5)
v_range = (v, -5, 5)

plot3d_parametric_surface(expr_x, expr_y, expr_z, u_range, v_range);
```

## Função Implícita

Função implícita de única variável

Utilizamos a função **plot_implicit(equação)** para desenhar gráficos de funções implícitas

```python
p1 = plot_implicit(Eq(x**2 + y**2-5))
```

# Função implícita com múltiplas variáveis

```python
p2 = plot_implicit(
    Eq(x**2 + y**2, 3),
    (x, -3, 3), 
    (y, -3, 3)
)
```

## Desigualdades implícita

Para utilizarmos, passamos uma desigualdade para a função **plot_implicit**

```python
plot_implicit(y > x**2);
```

Para combinarmos diversas condições para definir uma região, podemos utilizar as conjunções lógicas **And** e **Or**

```python
plot_implicit(And(y > x, y > -x));
```

```python
plot_implicit(Or(y > x, y > -x));
```

As vezes SymPy não escolhe a variável para o eixo horizontal como nós esperamos

```python
plot_implicit(Eq(y - 1));
```

Neste caso, utilizamos **x_var** para escolhermos a variável para o eixo-X

```python
plot_implicit(Eq(y - 1),x_var=x);
```
