# Visualizando uma Onda Senoidal com Matplotlib e NumPy

- Ondas senoidais representam oscilações periódicas
- Ondas senoidais possuem a forma de uma curval senoidal
- O eixo-X da curva senoidal representa tempo
- O eixo-Y da curva senoidal representa a amplitude da onda senoidal
- A amplitude da onda senoidal em qualquer ponto de Y é proporcional ao seno da variável
- A onda senoidal é dada pela equação:

\begin{equation}A\sin(\omega t)\end{equation}

Ao qual:

$A$ - Amplitude

$t$ - Tempo

$\omega$ - Frequência

- A curva senoidal passa pela origem (0,0)
- Um ciclo de Onda Senoidal é completado quando a posição da onda senoidal começa por uma posição e vai para a mesma posição após atingir sua amplitude máxima e mínima durante seu curso
- O tempo levado para completar um ciclo é chamado de período da Onda Senoidal
- A frequência da Onda Senoidal é dada pelo número de ciclos por segundo
- 'A' denota a amplitude da Onda Senoidal
- A distância coberta por um ciclo, mede o comprimento de onda da Onda Senoidal
- O comprimento de onda da Onda Senoidal é denotado por $\lambda$
- Exemplos de Ondas Senoidais incluem as oscilações produzidas por um peso suspenso em uma mola e a corrente alternada
- NumPy possui a função **sin()**, no qual recebe um array de valores e provê o valor do seno para eles
- Utilizando a função **sin()** de NumPy e a função **plot()** de matplotlib, podemos desenhar uma Onda Senoidal

```python
import numpy as np
import matplotlib.pyplot as plot
%matplotlib inline
plot.rcParams["figure.figsize"] = [16,9]
 
# Obtém os valores x da onda senoidal
tempo = np.arange(0, 10, 0.1);

# A amplitude da onda senoidal é o seno de uma variável, nesse caso será tempo
amplitude = np.sin(tempo)

# Plotamos a onda senoidal utilizando tempo e amplitude
plot.plot(tempo, amplitude)

# Damos um título ao Gráfico
plot.title('Onda Senoidal')

# Rotulamos o eixo-X
plot.xlabel('Tempo')

# Rotulamos o eixo-Y
plot.ylabel('Amplitude = sin(time)')
plot.grid(True, which='both')
plot.axhline(y=0, color='k')
plot.show()

# Mostra o gráfico da onda senoidal
plot.show()
```
