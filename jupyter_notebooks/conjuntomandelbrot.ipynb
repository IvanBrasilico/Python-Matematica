# O Conjunto de Mandelbrot

O Conjunto de Mandelbrot, nomeado em homenagem ao grande matemático Benoit Mandelbrot. Graças ao seu acesso aos computadores IBM, Mandelbrot foi um dos primeiros a utilizar computação gráfica para criar e exibir imagens geométricas fractais, levando ele a descoberta do Conjunto de Mandelbrot em 1980.

Ele demonstrou como a complexidade visual pode ser criada através de regras simples.

Fractais são objetos que exibem auto-similaridade em várias escalas. Ampliar um fractal revela detalhes de escala-menor similares às de escalar-maior. Embora o Conjunto de Mandelbrot seja auto-similar em escalas ampliadas, os detalhes de escala-menor não são idênticos ao todo. De fato, o Conjunto de Mandelbrot é infinitamente complexo. O processo de gerá-lo é baseado em uma equação extremamente simples envolvendo números complexos $(\mathbb{C})$. 

## Compreendendo os números complexos

O Conjunto Mandelbrot é um conjunto matemático, ou seja, uma coleção de números. Esses números são diferentes dos números reais que utilizamos em nosso dia-a-dia. Eles são **números complexos**. Um número complexo consiste de um **número real** mais um **número imaginário**. O número real é um número ordinário, por exemplo, $-2$. O número imaginário é um número real vezes um número especial chamado de $i$, por exemplo, $4i$. Um exemplo de um número complexo seria: $-2 + 4i$.

O número $i$ foi inventando pois nenhum número real pode ser elevado ao quadrado (multiplicado por si mesmo) e resultar em um número negativo. Isso significa que não podemos extrair a **raiz quadrada** de um número negativo e obter um número real. Por definição, quando extraímos a raiz quadrada de um número, nós encontramos um número que pode ser elevado ao quadrado para obter aquele número. O número $i$ é definido por ser a raíz quadrada de $-1$. Isso significa que $i$ ao quadrado é igual a $-1$, ou seja, quando nós elevamos um número imaginário ao quadrado nós podemos obter um número negativo. Por exemplo, $3i$ ao quadrado é igual a $-9$.

Números reais podem ser representados em uma linha de uma dimensão chamada **linha dos números reais**. Números negativos como -2 são plotados para a esquerda do 0 e números positivos são plotados para a direita do 0. Qualquer número real pode ser projetado na linha dos números reais

![img](http://www.ddewey.net/mandelbrot/realline.png)

Uma vez que os números complexos possuem duas partes, uma real e uma imaginária, nós precisamos de uma segunda dimensão pra podermos projetá-los. Para isso adicionamos uma dimensão vertical à linha de números reais que representará a parte imaginária. Uma vez que nosso gráfico agora é de duas dimensões, ele agora é um plano, o **plano dos números complexos**. Nós podemos projetar qualquer número complexo nesse plano. Veja o exemplo do número $3 + 4i$.

## Definição Formal do Conjunto de Mandelbrot

O Conjunto de Mandelbrot é um conjunto de valores de $c$ no plano complexo para qual a órbita de 0 sob iteração do mapa quadrático permanece limitado. 

\begin{equation}z_{n+1} = z_n^2 + c\end{equation}

Isto é, um número complex $c$ é parte do conjunto de Mandelbrot se, quando começamos com $z_0 = 0$ e aplicando a iteração repetidamente, o valor absoluto de $z_n$ permanece limitado não importa quão grande $n$ se torna. Isso também pode ser representado como:

\begin{equation}z_{n+1} = z_n^2 + c\end{equation}
\begin{equation}c \in M \iff \limsup_{n\to\infty}|z_{n+1}| \leqslant 2\end{equation}

Por exemplo, deixando $c = 1$ nos fornece a sequência $0, 1, 2, 5, 26, ...$, ao qual tende para o infinito. Como essa sequência não é limitada, 1 não é um elemento do Conjunto de Mandelbrot. Por outro lado, caso $c = -1$ nos fornece a sequência $0, -1, 0, -1, 0, ...$ no qual é limitada, então quer dizer que $-1$ pertence ao conjunto de Mandelbrot.

O Conjunto de Mandelbrot $M$ é definido por uma família de polinômios quadráticos complexos.

$P_c : \mathbb{C} \to \mathbb{C}$

Dado por:

$P_c : z \to z^2 + c$

Onde $c$ é um parâmetro complexo. Para cada $c$, considera-se o comportamento da sequência

$(0, P_c(0),P_c(P_c(0)),P_c(P_c(P_c(0))),...)$

Obtida pela iteração $P_c(z)$ começando pelo ponto crítico $z=0$, no qual escapa para a infinidade ou permanece dentro de um disco de algum raio finito. O conjunto de Mandelbrot é definido como o conjunto de todos os pontos $c$ tal que a sequência acima não fuja para a infinidade.

Analisando formalmente, se $P_c^n(z)$ denota o $n$ iterado de $P_c(z)$, o conjunto Mandelbrot é o subconjunto de plano complexo dado por

\begin{equation}M = \{c \in \mathbb{C} | (\exists s \in \mathbb{R}^+) (\forall n \in \mathbb{N}) (|P_c^n(0)| \leqslant s) ) \}\end{equation}

Como já explicado, é de fato possível simplificar essa definição ao assumirmos $s = 2$.

## Sumarizando

Matematicamente, o Conjunto de Mandelbrot é apenas um conjunto de números complexos. Dado um número complex $c$ ou ele pertencerá a $M$ ou ficará de fora. Uma pintura do conjunto de Mandelbrot pode ser feita, por exemplo, ao colorirmos todos os pontos $c$ que pertencem à $M$ da cor preta e todos os outros pontos de branco. 

As imagens do conjunto de Mandelbrot que possuem mais cores são geradas ao colorirmos pontos não pertencentes ao conjunto de acordo com qual termo na sequência $|P_c^n(0)|$ é o primeiro termo com um valor absoluto maior do que certo valor de corte, usualmente $2$, a lista de cores usada é sempre pré-definida pelo programa usado ou criada pelo usuário, a próxima cor da lista é escolhida quando o contador da iteração aumenta.

### Vejamos alguns exemplos da implementação do Conjunto de Mandelbrot

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import newaxis
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
plt.rcParams["figure.figsize"] = [18,11]

def compute_mandelbrot(N_max, some_threshold, nx, ny):
    # Uma grid de valores c
    x = np.linspace(-2, 1, nx)
    y = np.linspace(-1.5, 1.5, ny)
    c = x[:,newaxis] + 1j*y[newaxis,:]

    # Iteração de Mandelbrot
    z = c
    for j in range(N_max):
        z = z**2 + c

    mandelbrot_set = (abs(z) < some_threshold)
    return mandelbrot_set

mandelbrot_set = compute_mandelbrot(50, 50., 601, 401)

plt.imshow(mandelbrot_set.T, extent=[-2, 1, -1.5, 1.5])
plt.gray()
plt.show()
```

### Implementação com diversas cores

```python
from pylab import *
from numpy import NaN
 
def m(a):
    z = 0
    for n in range(1, 100):
        z = z**2 + a
        if abs(z) > 2:
            return n
    return NaN
 
X = arange(-2, .5, .002)
Y = arange(-1,  1, .002)
Z = zeros((len(Y), len(X)))
 
for iy, y in enumerate(Y):
    #print ("iteração", iy, "de", len(Y))
    for ix, x in enumerate(X):
        Z[iy,ix] = m(x + 1j * y)
 
imshow(Z, cmap = plt.cm.prism, interpolation = 'none', extent = (X.min(), X.max(), Y.min(), Y.max()))
xlabel("Re(c)")
ylabel("Im(c)")
savefig("mandelbrot_set.svg")
show()
```

### Utilizando a biblioteca PIL

```python
from PIL import Image 
from numpy import complex, array 
import colorsys 
  
# Seta a largura da imagem de output como 1024
WIDTH = 1024
  
# Uma função que retorna uma tupla de cores como valores inteiros de RGB
def rgb_conv(i): 
    color = 255 * array(colorsys.hsv_to_rgb(i / 255.0, 1.0, 0.5)) 
    return tuple(color.astype(int)) 
  
# Função que define Mandelbrot
def mandelbrot(x, y): 
    c0 = complex(x, y) 
    c = 0
    for i in range(1, 1000): 
        if abs(c) > 2: 
            return rgb_conv(i) 
        c = c * c + c0 
    return (0, 0, 0) 
  
# Criando a nova imagem no modo RGB
img = Image.new('RGB', (WIDTH, int(WIDTH / 2))) 
pixels = img.load() 
  
for x in range(img.size[0]): 
  
    # Exibindo o progresso como percentagem
    # print("%.2f %%" % (x / WIDTH * 100.0))  
    for y in range(img.size[1]): 
        pixels[x, y] = mandelbrot((x - (0.75 * WIDTH)) / (WIDTH / 4), 
                                      (y - (WIDTH / 4)) / (WIDTH / 4)) 
  
# Mostra o fractal criado após completar o número dado de iterações
img.show() 
```
