# Aplicando a Transformada Inversa de Fourier utilizando NumPy

- Enquanto que a Transformada de Fourier de Tempo Discreta transforma um sinal do domínio de tempo para o domínio de frequência, a Transformada Inversa de Fourier de Tempo Discreta recebe a representação do sinal de volta para o domínio de tempo
- A biblioteca NumPy possui uma função **ifft()** que executa a transformação inversa
- O exemplo a seguir utiliza uma onda senoidal com múltiplas frequências 1 Hertz, 2 Hertz e 4 Hertz.
- O sinal é plotado usando a função **numpy.fft.ifft()**

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.figsize"] = [16,9]

# Período de tempo
t = np.arange(0, 10, 0.01);

# Cria uma onda senoidal com múltiplas frequências(1 Hz, 2 Hz e 4 Hz)
a = np.sin(2*np.pi*t) + np.sin(2*2*np.pi*t) + np.sin(4*2*np.pi*t);

# Executa uma transformada de Fourier no sinal
tx = np.fft.fft(a);

# Executa uma transformada inversa de Fourier no sinal
itx = np.fft.ifft(tx);

# Plotamos a onda senoidal original utilizando a transformada inversa de Fourier
plt.plot(t, a);
plt.title("Gráfico da onda senoidal utilizando transformada inversa de Fourier");
plt.xlabel('Tempo')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show();
```
